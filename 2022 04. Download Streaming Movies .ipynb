## 2. With the increment sequence, download all files to a folder

<a href="C:\Users\Tat Sarkar\Downloads\Movie_Python"> Downloads >> Movie_ Python <a>


```python
def download_parts(link_pre, link_post , download_destination):
    import requests

    folder = download_destination

    for i in range(0,2500):
        link = link_pre+str(i+1)+link_post
        filename = "part_" + str(i+1).zfill(4) + ".ts"
        print (link)
        r = requests.get(link, allow_redirects=True)
        open(folder+ filename, 'wb').write(r.content)
        time.sleep(0.2)

download_destination = 'C:\\Users\\Tat Sarkar\\Downloads\\Movie_Python\\'
link_post = ".ts"

download_parts(link_pre, link_post , download_destination)
```

### 2B. I have now downloaded a lot of files. Need to delete files smaller than 1KB

To be safe, I downloaded a lot of files. But, now, I need to clean up the ones which are just empty *.ts files. I will do it by checking the size and deleting. 

```python
import os
import glob

for input in glob.glob( download_destination + "*.ts"):
    if(os.path.getsize(input) < 300): 
        os.remove(input)
```
